```python
import pandas as pd
import requests
import os
import random
from PIL import Image
from io import BytesIO
```

```python
file_path = '../csv_files/metadata_v2.csv'

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(file_path)
```

## Converting the links in METADATA_URL to obtain the original size image links for wga links

```python
'https://www.wga.hu/html/a/aertsen/apostles.html'
```

```python
'https://www.wga.hu/art/a/aertsen/apostles.jpg'
```

```python
link_df = df[df['IMAGE_LINK'].fillna('').str.contains('wga', case=False)].copy()

def convert_metadata_url(metadata_url):
    # Replace the URL path and file extension
    image_link = metadata_url.replace('/html', '/art').replace('.html', '.jpg')
    return image_link

# Apply the conversion function to create the 'HQ_IMAGE_LINK' column
link_df['IMAGE_LINK'] = link_df['METADATA_URL'].apply(convert_metadata_url)

df.update(link_df['IMAGE_LINK'])
```

```python
filtered_df = df[df['TECHNIQUE'].fillna('').str.contains('oil', case=False)]
len(filtered_df)
```

## Downloading 5 randomly selected Images from filtered_df

```python
# For reproducibility
random.seed(10)

# Shuffling and taking first five rows
df_shuffled = filtered_df.sample(frac=1, random_state=10)
selected_rows = df_shuffled.iloc[:5]

# Create directory
os.makedirs('oiloncanvas/', exist_ok=True)

# Download and save the selected images
for _, row in selected_rows.iterrows():
    image_link = row['HQ_IMAGE_LINK']
    title = row['TITLE']
    technique = row['TECHNIQUE']
    file_name = f"oiloncanvas/{title}_{technique}.jpg"
    

    response = requests.get(image_link)
    
    # Saving Image
    with open(file_name, 'wb') as file:
        file.write(response.content)
        print(f"Image saved: {file_name}")
```

```python
def verify_image_url(url):
    try:
        response = requests.get(url)
        image = Image.open(BytesIO(response.content))
        image.verify()
        return True
    except Exception as e:
        print(f"Error for link : {str(e)}")
        return False

selected_rows = filtered_df

    
# Download and save the selected images
for _, row in selected_rows.iterrows():
    image_url = row['IMAGE_LINK']
    
    # Verifying Link
    if verify_image_url(image_url):
        print(f'{_}')
    else:
        print(f"Link {url}")
```
